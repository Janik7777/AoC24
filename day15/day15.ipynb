The lanternfish use their own custom Goods Positioning System (GPS for short) to track the locations of the boxes. The GPS coordinate of a box is equal to 100 times its distance from the top edge of the map plus its distance from the left edge of the map. (This process does not stop at wall tiles; measure all the way to the edges of the map.)

So, the box shown below has a distance of 1 from the top edge of the map and 4 from the left edge of the map, resulting in a GPS coordinate of 100 * 1 + 4 = 104.

```
#######
#...O..
#......
```

The lanternfish would like to know the sum of all boxes' GPS coordinates after the robot finishes moving. In the larger example, the sum of all boxes' GPS coordinates is 10092. In the smaller example, the sum is 2028.

Predict the motion of the robot and boxes in the warehouse. After the robot is finished moving, what is the sum of all boxes' GPS coordinates?

````python
def result(inputfile):

    map = []
    moves = ""
    readingMap = True
    with open(inputfile) as f:
        for line in f:
            if readingMap:
                if len(line) == 1:
                    readingMap = False
                    continue
                tiles = [*line.strip()]
                map.append(tiles)
            else:
                moves += line.strip()
    
    for i, row in enumerate(map):
        for j, value in enumerate(row):
            if value == '@':
                y_robot = i
                x_robot = j 

    def move(direction, x, y):
        match direction:
            case 'v':
                next_x = x
                next_y = y + 1
                nextTile = map[next_y][next_x]
            case '^':
                next_x = x
                next_y = y - 1
                nextTile = map[next_y][next_x]
            case '>':
                next_x = x + 1
                next_y = y
                nextTile = map[next_y][next_x]
            case '<':
                next_x = x - 1
                next_y = y
                nextTile = map[next_y][next_x]
        
        if nextTile == '#':
            return False
        elif nextTile == '.':
            map[next_y][next_x] = map[y][x]
            map[y][x] = '.'
        else:
            move_possible = move(direction, next_x, next_y)
            if not move_possible:
                return False
            else:
                map[next_y][next_x] = map[y][x]
                map[y][x] = '.'
        
        return True

    for direction in moves:
        if move(direction, x_robot, y_robot):
            match direction:
                case 'v':
                    y_robot += 1
                case '^':
                    y_robot -= 1
                case '>':
                    x_robot += 1
                case '<':
                    x_robot -= 1
    
    result = 0
    for i, row in enumerate(map):
        for j, value in enumerate(row):
            if value == 'O':
                result += 100 * i + j
    
    return result
````

This warehouse's layout is surprisingly similar to the one you just helped. There is one key difference: everything except the robot is twice as wide! The robot's list of movements doesn't change.

To get the wider warehouse's map, start with your original map and, for each tile, make the following changes:

- If the tile is #, the new map contains ## instead.

````python
def result2(inputfile):

    map = []
    moves = ""
    readingMap = True
    with open(inputfile) as f:
        for line in f:
            if readingMap:
                if len(line) == 1:
                    readingMap = False
                    continue
                tiles = ""
                for char in line.strip():
                    match char:
                        case '#':
                            tiles += "##"
                        case 'O':
                            tiles += "[]"
                        case '.':
                            tiles += ".."
                        case '@':
                            tiles += "@."


                map.append([*tiles])
            else:
                moves += line.strip()

    for i, row in enumerate(map):
        for j, value in enumerate(row):
            if value == '@':
                y_robot = i
                x_robot = j

    def moveLeftRight(direction, x, y):
        match direction:
            case '>':
                next_x = x + 1
                next_y = y
                nextTile = map[next_y][next_x]
            case '<':
                next_x = x - 1
                next_y = y
                nextTile = map[next_y][next_x]
        
        if nextTile == '#':
            return False
        elif nextTile == '.':
            map[next_y][next_x] = map[y][x]
            map[y][x] = '.'
        else:
            move_possible = moveLeftRight(direction, next_x, next_y)
            if not move_possible:
                return False
            else:
                map[next_y][next_x] = map[y][x]
                map[y][x] = '.'
        
        return True
    

    def moveUpDownPossible(direction, x, y):
        match direction:
            case 'v':
                next_x = x
                next_y = y + 1
                nextTile = map[next_y][next_x]
            case '^':
                next_x = x
                next_y = y - 1
                nextTile = map[next_y][next_x]
        
        if nextTile == '#':
            return False
        elif nextTile == '.':
            return True
        elif nextTile == '[':
            return moveUpDownPossible(direction, next_x, next_y) and moveUpDownPossible(direction, next_x + 1, next_y)
        else:
            return moveUpDownPossible(direction, next_x, next_y) and moveUpDownPossible(direction, next_x - 1, next_y)
    
    def moveUpDown(direction, x, y):
        match direction:
            case 'v':
                next_x = x
                next_y = y + 1
                nextTile = map[next_y][next_x]
            case '^':
                next_x = x
                next_y = y - 1
                nextTile = map[next_y][next_x]
        if nextTile == '.':
            map[next_y][next_x] = map[y][x]
            map[y][x] = '.'
        elif nextTile in ['[', ']']:
            if nextTile == '[':
                moveUpDown(direction, next_x, next_y)
                moveUpDown(direction, next_x + 1, next_y)
            else:
                moveUpDown(direction, next_x, next_y)
                moveUpDown(direction, next_x - 1, next_y)

            map[next_y][next_x] = map[y][x]
            map[y][x] = '.'



    for direction in moves:
        if direction in ['<', '>']:
            if moveLeftRight(direction, x_robot, y_robot):
                match direction:
                    case '>':
                        x_robot += 1
                    case '<':
                        x_robot -= 1
        else:
            if moveUpDownPossible(direction, x_robot, y_robot):
                moveUpDown(direction, x_robot, y_robot)
                match direction:
                    case 'v':
                        y_robot += 1
                    case '^':
                        y_robot -= 1
    
    result = 0
    for i, row in enumerate(map):
        for j, value in enumerate(row):
            if value == '[':
                result += 100 * i + j
    
    return result
````
